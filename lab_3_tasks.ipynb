```python
print("""
    Task-9
""")

def most_frequent(st):
    t = dict()
    l = list()
    for i in st:
        d = st.count(i)
        if i not in t:
            t.update({ i : d})
    for key in t:
        l.append((t[key], key))
    l.sort(reverse = True)
    return l
s = most_frequent('rishi')
print(s)
```
